```python
import json

INPUT_FILE = "links_result.json"
OUTPUT_FILE = "links_result_deduped.json"

with open(INPUT_FILE, "r") as f:
    data = json.load(f)

# Step 1: Flatten - collect every link (keys + all values) into one single list
flat_links = []
for parent, links in data.items():
    flat_links.append(parent)
    if isinstance(links, list):
        flat_links.extend(links)

# Step 2: Delete duplicates - keep first occurrence only, preserve order
seen = set()
unique_links = []
for link in flat_links:
    if link not in seen:
        seen.add(link)
        unique_links.append(link)

with open(OUTPUT_FILE, "w") as f:
    json.dump(unique_links, f, indent=2)

print(f"Total links before dedup: {len(flat_links)}")
print(f"Total unique links: {len(unique_links)}")
print(f"Saved to {OUTPUT_FILE}")
```
